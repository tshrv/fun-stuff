```python
import random
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
n = 100000
```

```python
primary_vertices = ((0, 0), (0.25, 1), (1, 0.4))
seed = (random.random(), random.random())
print(seed)
generated_points = {'x': [], 'y': []}
```

```python
for _ in range(n):
    x1, y1 = random.choice(primary_vertices)    # point_1
    x2, y2 = seed   # point_2
    x3, y3 = min(x1, x2) + abs(x2-x1)/2, min(y1, y2) + abs(y2-y1)/2
    generated_points['x'].append(x3)
    generated_points['y'].append(y3)
    seed = (x3, y3)
```

```python
# generated_points
```

```python
generated_points['x'].extend(map(lambda x: x[0], primary_vertices))
generated_points['y'].extend(map(lambda x: x[1], primary_vertices))
```

```python
g = sns.relplot(
    # data={'x': [0,.5,1], 'y': [0,1,0]},
    data=generated_points,
    x="x", y="y",
    s=1
)
```
